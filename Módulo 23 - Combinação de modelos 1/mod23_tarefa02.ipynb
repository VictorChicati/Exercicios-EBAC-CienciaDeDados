#2. Explique com suas palavras o random forest


---

O random forest pode ser explicado como sendo um _update_ do bagging, oferecendo melhorias de desempenho e resultados, ele assim como o bagging combina modelos de Machine Learning em que o modelo é treinado em um conjunto de dados da base original, porém, ele é treinado com uma quantidade determinada e aleatória de variáveis, diminuindo a chance de overfitting.

---

#3. Qual a diferença entre bagging e random forest?


---

Em resumo, enquanto Bagging se concentra em criar modelos diversificados usando amostragem bootstrap, Random Forest vai além ao introduzir aleatoriedade na seleção de características, o que muitas vezes resulta em um desempenho melhor e uma maior resistência ao overfitting.
