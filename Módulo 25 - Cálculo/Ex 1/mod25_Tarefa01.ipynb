# Cálculo - Tarefa 01

1. Marque quais desses métodos/algoritmos muito populares em ciência de dados são baseados no uso de derivada:

    1. **Método Mínimos Quadrados**
    2. **Gradiente descendente**
    3. **Newton Raphson**
    4. CART (Árvore de decisão)


1. **Método Mínimos Quadrados:**
   - **Baseado no uso de derivada?** Sim. O método dos Mínimos Quadrados envolve a minimização da soma dos quadrados dos resíduos, e a solução analítica para os coeficientes requer a derivada em relação aos parâmetros.

2. **Gradiente Descendente:**
   - **Baseado no uso de derivada?** Sim. O Gradiente Descendente é um algoritmo de otimização que utiliza a derivada (gradiente) da função de custo em relação aos parâmetros para encontrar o mínimo local da função.

3. **Newton-Raphson:**
   - **Baseado no uso de derivada?** Sim. O método de Newton-Raphson é um algoritmo iterativo para encontrar raízes de uma função e usa a derivada da função para ajustar a aproximação da raiz.

4. **CART (Árvore de Decisão):**
   - **Baseado no uso de derivada?** Não. O algoritmo CART (Classification and Regression Trees) não é baseado no uso direto de derivadas. Ele utiliza métodos de particionamento recursivo, como a busca da melhor divisão baseada em critérios como Gini impurity ou redução de variância, sem envolver derivadas diretamente.


2. Dada uma base de dados com uma variável resposta $y$ e um conjunto de variáveis explicativas. Considere uma estrutura de um modelo de regressão. Explique com suas palavras por que não é possível obter parâmetros que forneçam um erro quadrático médio (EQM) menor que o obtido com estimadores de mínimos quadrados.

Os estimadores de mínimos quadrados são escolhidos para minimizar a soma dos quadrados dos resíduos, representando uma solução analítica única e eficiente na minimização da variância dos parâmetros em um modelo linear. Tentar obter parâmetros diferentes que proporcionem um erro quadrático médio (EQM) menor que os estimadores de mínimos quadrados implicaria violar a propriedade de otimalidade desses estimadores. Em condições ideais e sob as premissas do modelo linear clássico, não é possível obter uma solução que supere a eficiência dos mínimos quadrados. Qualquer tentativa nesse sentido só resultaria em um aumento do EQM, a menos que haja uma alteração nas premissas do modelo ou uma violação das suposições subjacentes. Isso destaca a robustez e a eficácia dos estimadores de mínimos quadrados no contexto de modelos lineares.
