

# Tarefa I

1. Qual a principal diferença entre um *behaviour score* e um *credit score*?

- **Behaviour score**: Refere-se à avaliação do comportamento de crédito atual de um cliente com base no histórico recente de transações e padrões de pagamento. É utilizado para monitorar e prever o comportamento futuro de crédito de clientes já existentes.
- **Credit score**: É uma pontuação geral que avalia a probabilidade de um indivíduo cumprir suas obrigações financeiras com base em diversos fatores, incluindo histórico de crédito, nível de endividamento, duração do histórico de crédito e tipos de crédito utilizados. É frequentemente usado para avaliar novos solicitantes de crédito.

2. Por que é tão comum se estabelecer janelas de desempenho de 12 meses, e 12 safras para o desenvolvimento de modelos?

- Estabelecer janelas de desempenho de 12 meses é comum porque permite capturar um ciclo completo de comportamento financeiro, incluindo variações sazonais que podem afetar o desempenho de crédito. Utilizar 12 safras, ou períodos, para o desenvolvimento de modelos ajuda a garantir que o modelo seja robusto e capaz de lidar com diferentes condições econômicas e comportamentais ao longo do tempo.

3. Qual a diferença entre amostra *out of time* e amostra *out of sample*?

- Out of time (OOT): Refere-se a uma amostra de dados que foi coletada em um período de tempo diferente daquele usado para treinar o modelo. É usada para avaliar a capacidade do modelo de generalizar para dados futuros.
- Out of sample (OOS): Refere-se a uma amostra de dados que não foi utilizada no processo de treinamento do modelo, mas que foi coletada durante o mesmo período de tempo. É usada para validar a performance do modelo em dados não vistos.

4. Se os dados da variavel resposta estão sistematicamente corrompidos para duas das safras de desenvolvimento, o que você faria?

- Se os dados da variável resposta estão corrompidos, é importante primeiro tentar identificar a causa e corrigir os dados, se possível. Se a correção não for viável, essas safras podem precisar ser excluídas do desenvolvimento do modelo para evitar que os dados corrompidos introduzam viés ou reduzam a precisão do modelo. Alternativamente, métodos de imputação de dados, podendo ser feito através de vários métodos, como a média, mediana, ou até mesmo modelos de machine learning mais complexos.
