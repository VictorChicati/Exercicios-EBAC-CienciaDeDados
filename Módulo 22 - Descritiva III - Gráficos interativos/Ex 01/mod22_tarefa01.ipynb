#### 1. Baixando dados de ativos

```python
import pandas as pd
from datetime import datetime


from pandas_datareader import data as pdr
import yfinance as yf
yf.pdr_override()
import plotly.express as px
```

```python
# obtendo a data de hoje
hoje = datetime.now().strftime("%Y-%m-%d")
hoje
```

```python
# Definir ticker e período desejado
ticker = ['GOOGL', 'PETR4.SA', 'MGLU3.SA']
inicio = '2019-01-01'
fim = hoje

# Obter dados históricos
dados_hist = pdr.get_data_yahoo(ticker, start=inicio, end=fim)
dados_hist.head()
```

```python
dados_hist.to_csv('dados_historicos_acoes.csv')
```

#### 2. Observando a volatilidade dentro do dia
1. Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).
2. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.
3. Veja se consegue identificar que ativo apresentou maior volatilidade.

```python
# carregando os dados
df = pd.read_csv('/content/dados_historicos_acoes.csv',
                 header= [0,1],
                 index_col=0)
```

```python
# juntando os dados dentro da mesma data
df = df.stack()
df
```

```python
df['Range'] = df['High'] - df['Low']
df = df.unstack()
df['Range']
```

```python
df_range = df['Adj Close'].stack().reset_index()
df_range.columns = ['data', 'ativo', 'valor']
df_range
```

```python
fig = px.line(df_range, x='data', y='valor', color='ativo', template='plotly_dark')
fig.show()
```

Podemos verificar que a empresa que obteve maior diferença no período observado foi a empresa 'GOOGL' que se trata da google
