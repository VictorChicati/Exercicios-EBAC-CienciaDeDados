#### 2. Média móvel

1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
2. Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

````python
# realizando a média janela móvel
df_mw = df['Adj Close'].rolling(window=90)
df_mw.mean()
````

````python
# realizando o desvio padrão da janela móvel
df_mw.std()
````

````python
#realizando alguns ajustes na janela móvel
df_media_movel = df_mw.mean().stack().reset_index()
df_media_movel.columns = ['Date', 'Ticker','Price']
df_media_movel
````

````python
#plotando o gráfico de média
fig = px.line(data_frame=df_media_movel,
              x='Date',
              y='Price',
              color='Ticker',
              template='plotly_dark')
fig.show()
````

````python
df_std_adjclose = df_mw.std().stack().reset_index()
df_std_adjclose.columns = ['data', 'ativo', 'valor']
#plotando o gráfico
fig = px.line(data_frame=df_std_adjclose,
              x='data',
              y='valor',
              color='ativo',
              template='plotly_dark')
fig.show()
````

Analisando ambos os gráficos de desvio padrão e média móvel, podemos verificar que o ativo com maior

#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
#selecionando os ultimos 60 valores
df_60 = df.xs(key='PETR4.SA', level=1, axis=1).tail(n=60)
df_60.head()
````

````python
# gerando um gráfico
layout = Layout(paper_bgcolor='rgba(0,0,0,0)',
                plot_bgcolor='black',
                height=600)

graph = {
    'x': df_60.index,
    'open': df_60.Open,
    'close': df_60.Close,
    'high': df_60.High,
    'low': df_60.Low,
    'type': 'candlestick',
    'name': 'PETR4.SA',
    'showlegend': True
}

grafico = graph_objects.Figure(data=[graph], layout_title='', layout=layout)

grafico.update_xaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')
grafico.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')

grafico
````
