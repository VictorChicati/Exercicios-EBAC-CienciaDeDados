```python
# Criando uma matriz do Patsy com as variáveis qualitativas e com a casela de referencia sendo a categoria mais frequente
y, X = patsy.dmatrices('np.log(renda) ~ data_ref + C(sexo, Treatment("F")) + posse_de_veiculo + posse_de_imovel + qtd_filhos + C(tipo_renda, Treatment("Assalariado")) + C(educacao, Treatment("Secundário")) + C(estado_civil, Treatment("Casado")) + C(tipo_residencia, Treatment("Casa"))  + idade + tempo_emprego + qt_pessoas_residencia', data = df)
```

```python
mod = sm.OLS(y,X).fit()
mod.summary()
```

Em uma análise rápida observando os p-values, podemos concluir que algumas das variáveis não tem significancia para o nosso modelo, sendo elas:

 - Algumas datas de referencia
 - Renda bolsista
 - Renda Pensionista
 - Em educação apenas a categoria "superior completo" possui p-value abaixo de 5%, porém, como o valor é 0%, precisamos realizar uma análise mais detalhada para entender melhor
 - Estado civil: União
 - Tipo de residencia




```python
# Removendo a variável menos significantes: data_ref
y, X = patsy.dmatrices('np.log(renda) ~ C(sexo, Treatment("F")) + posse_de_veiculo + posse_de_imovel + qtd_filhos + C(tipo_renda, Treatment("Assalariado")) + C(educacao, Treatment("Secundário")) + C(estado_civil, Treatment("Casado")) + C(tipo_residencia, Treatment("Casa"))  + idade + tempo_emprego + qt_pessoas_residencia', data = df)
mod = sm.OLS(y,X).fit()
mod.summary()
```

Ao realizar a retirada da variável descrita acima, temos então um modelo com um r-squared ligeiramente menor que o inicial, podendo demonstrar uma piora da acurácia do modelo

```python
# Removendo outras variáveis não significantes: educacao, tipo_residencia, estado_civil
y, X = patsy.dmatrices('np.log(renda) ~  C(sexo, Treatment("F")) + posse_de_veiculo + posse_de_imovel + qtd_filhos + idade + tempo_emprego + qt_pessoas_residencia', data = df)
mod = sm.OLS(y,X).fit()
mod.summary()
```

```python
# Retirando a variável qt_pessoas_residencia
y, X = patsy.dmatrices('np.log(renda) ~  C(sexo, Treatment("F")) + posse_de_veiculo + posse_de_imovel + qtd_filhos + idade + tempo_emprego', data = df)
mod = sm.OLS(y,X).fit()
mod.summary()
```

Agora, temos apenas interceptos com um valor abaixo de 5%. O r-squared se encontra menor que o do primeiro modelo, o que pode nos mostrar uma menor aderência do modelo
